```python
input_string = input("Enter a string: ")
cleaned_string = input_string.lower().replace(" ", "")

reversed_string = cleaned_string[::-1]

if cleaned_string == reversed_string:
    print(f"The string '{input_string}' is a palindrome.")
else:
    print(f"The string '{input_string}' is not a palindrome.")
```

```python
# write a prgrm to print evry 2nd number from 0 to 10 using while loop
num = 0
while num <= 10:
    if num % 2 == 0:
        print(num)
    num += 1
```

```python
# write a programe to print all even numbers between 2 and 20

num = 2
while num <= 20:
    if num % 2 == 0:
        print(num,end=' ')
    num += 2
```

```python
# write a programe to reverse a number using while loop
num = int(input("Enter a number to reverse: "))
reversed_num = 0

while num > 0:
    remainder = num % 10
    reversed_num = (reversed_num * 10) + remainder
    num = num // 10

print(f"The reversed number is: {reversed_num}")
```
